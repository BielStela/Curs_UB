While PCA is flexible, fast, and easily interpretable, it does not perform so well when there are _nonlinear_ relationships within the data.

```python
%matplotlib inline
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA
```

```python
X, y = make_swiss_roll(n_samples=3000, noise=0.05)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Scatter3d(
    x=X[:,0],
    y=X[:,1],
    z=X[:,2],
    mode='markers',
    marker=dict(
        size=5,
        color=y,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)

data = [trace1]
layout = go.Layout(margin=dict(l=0,r=0,b=0,t=0))
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='3d-scatter-colorscale')
```

```python
pca = PCA()
x_pca = pca.fit_transform(X)
x_pca.shape
```

```python
plt.scatter(x_pca[:,0], x_pca[:,1], c=y, cmap='viridis')
```

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
x_lle = lle.fit_transform(X)
```

```python
plt.scatter(x_lle[:,0], x_lle[:,1], c=y, cmap='viridis')
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE()
x_tsne= tsne.fit_transform(X)
```

```python
plt.scatter(x_tsne[:,0], x_tsne[:,1], c=y, cmap='viridis')
```

### UMAP

```python
import umap
```

```python
um = umap.UMAP(n_neighbors=50)
x_umap = um.fit_transform(X)
```

```python
plt.scatter(x_umap[:,0], x_umap[:,1], c=y, cmap='viridis')
```

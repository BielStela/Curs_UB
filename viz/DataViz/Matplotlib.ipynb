# Matplotlib

Is the biggest and most used plotting library in the Python ecosystem. Matplotlib is a multiplatform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack.

It is extremely versatile and supports dozens of backends and output formats. In recent years, however, the interface and style of Matplotlib have begun to show their age. Compared to newer tools like ggplot in the R language or web visualization toolkits based on D3js and HTML5 canvas, often make Matplotlib feel clunky and old-fashioned. 

But thanks to the open community, there are lots of high level libraries that make use of the powerful matplotlib engine and give a clean interface to build plots with simple commands (Seaborn, Pandas, HoloViews...).

In order to use those high levels libraries, it will be handy to understand the matplotlib internals first.

# The interfaces

A potentially confusing feature of Matplotlib is its dual interfaces: a convenient
MATLAB-style state-based interface, and a more powerful object-oriented interface.
We’ll quickly highlight the differences between the two here

## MATLAB style interface

The MATLAB-style tools are contained in the pyplot (`plt`) interface. `plt` interface give use functions that "activate" a figure and every `plt` statement modifies the current "active" figure.

```python
plt.plot(np.random.randn(10), np.random.randn(10), 'bo')
```

To create multiple plots in the same figure:

```python
x = np.linspace(0,10,70)  # data

# defines a subplot in the given position
plt.subplot(2, 1, 1) # nrows, ncols, index
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.tight_layout()
```

If you want to put the two lines in the same axis just do:

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

or **looping**:

```python
for i in range(2):
    plt.plot(x, np.sin(x+i*(np.pi/2)))
```

```python
plt.scatter(x, np.sin(x))
```

## Object oriented style interface

The object-oriented interface is available for these more complicated situations, and
for when you want more control over your figure.

It can be used in several ways:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)  # just one argument -> n rows of axes

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

or:

```python
# first create the figure object
fig = plt.figure()

# initialize the axes objest
ax1 = fig.add_subplot(2,1,1)
ax1.plot(x, np.sin(x))

# initialize a second axes object at the next position of the grid
ax2 = fig.add_subplot(2,1,2)
ax2.plot(x, np.cos(x))
```

Depending on the complexity of the plot we want to achieve, one of these approaches is the convenient

## Plot customization
### Line style


```python
x = np.arange(10)
plt.plot(x, x + 1, linestyle='-', linewidth=4) # solid
plt.plot(x, x + 2, linestyle='--') # dashed
plt.plot(x, x + 3, linestyle='-.') # dashdot
plt.plot(x, x + 4, linestyle=':'); # dotted
```

### Marker style

```python
data = [-1, 5, 3, 7, 6, 8]
plt.plot(data)
```

```python
plt.plot(data, marker='o')
```

```python
plt.plot(data, marker='o', linestyle='--')
```

```python
plt.plot(data,color='gray', linestyle='--', marker='o', markerfacecolor='r')
```

### adjusting the axes

```python
plt.plot(data,color='gray', linestyle='--', marker='o', markerfacecolor='k')
plt.xlim(-1, 11)
plt.ylim(-1.5, 10)
```

```python
plt.plot(data,color='gray', linestyle='--', marker='o', markerfacecolor='k')
plt.axis('equal');
```

```python
plt.plot(data,color='gray', linestyle='--', marker='o', markerfacecolor='b')
plt.axis('off');
```

but to customize further we need to switch to the OO interface with `plt.gca()` which means "get current axis".
The "lines" of the axis are calle "spines" and are via `ax.spines['position']`.

```python
plt.plot(data,color='gray', linestyle='--', marker='o', markerfacecolor='b')
plt.xlim(-5, 8)
plt.ylim(-1.5, 10)

ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')  # or set_visible(False)
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

### Scatter with lots of points

```python
x1 = np.random.randn(1000, 2)
x2 = np.random.normal([10,10], 5, (5000,2))
x = np.vstack((x1, x2))
```

```python
plt.scatter(x)  # scatter plots need X and Y
```

```python
plt.scatter(x[:,0], x[:,1])
```

How can we visualize this data?

+ add alpha:

```python
plt.scatter(x[:,0], x[:,1], alpha=0.1)
```

+ make the points small

```python
plt.scatter(x[:,0], x[:,1], s=1)
```

+ use hist2d

```python
p = plt.hist2d(x[:,0], x[:,1])
```

```python
plt.hist2d(x[:,0], x[:,1], bins=20);
```

two visualize better the distirbution we could use a [log color map](https://matplotlib.org/users/colormapnorms.html)

```python
plt.figure(figsize=(10,10))
plt.hexbin(x[:,0], x[:,1], gridsize=50)
```

## Plot legend and annotation
define the label of every element with `label=` argument

```python
x = np.linspace(-5,5,100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
```

```python
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
```

```python
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
